```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder


df = pd.read_csv("sampled_data.txt")
df
```

```python
df.value_counts('shotType')
```

```python
# Encoding shotType column

encoder = OneHotEncoder(sparse_output=False)
print(df['location'].value_counts())
encoded_data = encoder.fit_transform(df['location'].values.reshape(-1, 1))

encoded_df = pd.DataFrame(encoded_data, columns = ["HOMEZONE", "AWAYZONE", "NEUTRALZONE"])

print(encoded_df)
df.insert(10, "HOMEZONE", encoded_df["HOMEZONE"])
df.insert(11, "AWAYZONE", encoded_df["AWAYZONE"])
df.insert(12, "NEUTRALZONE", encoded_df["NEUTRALZONE"])
df.drop("location", axis=1, inplace=True)
df
```

```python
df['shotType'].value_counts()
encoder = OneHotEncoder(sparse_output=False)
encoded_data = encoder.fit_transform(df['shotType'].values.reshape(-1, 1))

encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['shotType']))
df.insert(25, "shotType_BACK", encoded_df["shotType_BACK"])
df.insert(26, "shotType_DEFL", encoded_df["shotType_DEFL"])
df.insert(27, "shotType_SLAP", encoded_df["shotType_SLAP"])
df.insert(28, "shotType_SNAP", encoded_df["shotType_SNAP"])
df.insert(29, "shotType_TIP", encoded_df["shotType_TIP"])
df.insert(30, "shotType_WRAP", encoded_df["shotType_WRAP"])
df.insert(31, "shotType_WRIST", encoded_df["shotType_WRIST"])

df.drop("shotType", axis=1, inplace=True)
```
